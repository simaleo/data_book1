```python
# Importation des librairies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
pd.options.display.max_columns = None
pd.options.display.max_rows = None
warnings.filterwarnings('ignore')
```

```python
# Importation des données depuis mon sit Github
file = "https://raw.githubusercontent.com/simaleo/data_book1/main/hdv2003.csv"
data = pd.read_csv(file)
```

# REPRÉSENTATION GRAPHIQUE D’UNE VARIABLE QUALITATIVE

### DIAGRAMME CIRCULAIRE OU « CAMEMBERT »

```python
# Représentation circulaire de la variable occup
plt.pie(data['occup'].value_counts().values,
       labels = data['occup'].value_counts().index,
       autopct='%.2f%%')
plt.title("Diagramme circulaire ou camembert")
plt.show()
```

### DIAGRAMME EN BARRE

```python
# Représentation en barre de la variable occup
plt.subplots(figsize = (14, 4))
plt.bar(data['occup'].value_counts().index,
        data['occup'].value_counts().values) # Utilisation de la librairie matplotlib
plt.title("Diagramme en barre") # Titre du graphique
plt.xlabel("Occupations") # Titre de l'axe des abscisses
plt.ylabel('Effectif') # Titre de l'axe des ordonnées
plt.show()
```

```python
# Représentation en barre de la variable occup
plt.subplots(figsize = (14, 4))
sns.countplot(x = 'occup', data = data) # Utilisation de la librairie seaborn
plt.title("Diagramme en barre") # Titre du graphique
plt.xlabel("Occupations") # Titre de l'axe des abscisses
plt.ylabel('Effectif') # Titre de l'axe des ordonnées
plt.show()
```

# REPRÉSENTATION GRAPHIQUE D’UNE VARIABLE QUANTITATIVE

### HISTOGRAMME

```python
# Histogramme de la variable poids
sns.displot(x = 'poids', data = data)
plt.show()
```

```python
# Densité de probabilité de la variable poids
sns.displot(x = 'poids', kind = 'kde', data = data)
plt.show()
```

```python
# Histogramme et densité de probabilité de la variable poids
sns.displot(x = 'poids', kde = True, data = data)
plt.show()
```

### BOITE À MOUSTACHES

```python
# Boîte à moustaches de la variable poids
sns.boxplot(y = 'poids', data = data)
plt.title("Poids des individus")
plt.show()
```
