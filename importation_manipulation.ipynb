```python
# Exemple 1: On souhaite avoir seulement les informations du genre féminin
data[data['sexe'] == 'Femme'].head()
```

```python
# Exemple 3 : On souhaite avoir seulement les informations des personnes
#     qui ont un poids inférieur ou égal à 5731.66 g
data[data['poids'] <= 5731.66].head()
```

```python
# Exemple 1 : On souhaite avoir les informations des hommes qui ont plus de 25 ans
data[(data["sexe"] == 'Homme') & (data['age'] > 25)].head()
```

```python
# Exemple 2 : On souhaite avoir les informations des personnes qui exercent une profession et qui font du sport
data[(data["occup"] == "Exerce une profession") & (data['sport'] == 'Oui')].head()
```

```python
# Exemple 3 : On souhaite avoir les informations des personnes qui pratiquent régulièrement la religion ou 
# qui n'ont ni croyance ni appartenance
data[(data["relig"] == "Pratiquant regulier") | (data['relig'] == "Ni croyance ni appartenance")].head()
```

```python
# Effectif selon le genre des personnes faisant du sport
pd.crosstab(data['sexe'], data['sport'])
```

```python
# Niveau d’étude en fonction de l'appartence à une religion 
pd.crosstab(data["nivetud"], data["relig"])
```

```python
# Pratique du sport selon l’occupation des individus
pd.crosstab(data['sport'], data['occup'])
```

```python
# Pratique du sport selon l’occupation des individus
# Pourcentage en ligne
pd.crosstab(data['sport'], data['occup'], normalize='index')* 100
```

```python
# Pratique du sport selon l’occupation des individus
# Pourcentage en colonne
pd.crosstab(data['sport'], data['occup'], normalize='columns')* 100
```

```python
# Pratique du sport selon l’occupation des individus et le genre
# Pourcentage en ligne
pd.crosstab([data['sexe'], data['sport']], data['occup'], normalize='index')* 100
```

```python
# Pratique du sport selon l’occupation des individus et le genre
# Pourcentage en colonne
pd.crosstab([data['sexe'], data['sport']], data['occup'], normalize='columns')* 100
```

```python
# Représentation graphique par effectif

# Importation des librairies
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplots(figsize = (14, 4), dpi = 150)
sns.countplot(x = "occup", hue = 'sport', data = data)
plt.ylabel('Effectifs', fontsize = 12, labelpad = 16)
plt.xlabel('Occupation', fontsize = 12, labelpad = 16)
plt.show()
```

```python
# Représentation graphique par pourcentage en colonne

fig, axe = plt.subplots(figsize = (14, 4), dpi = 150)
dr = pd.crosstab(data['occup'], data['sport'], normalize='columns')* 100
dr.plot.bar(ax = axe)
plt.ylabel('Pourcentage(%)', fontsize = 12, labelpad = 16)
plt.xlabel('Occupation', fontsize = 12, labelpad = 16)
plt.xticks(rotation = 0)
plt.show()
```

```python
# Age minimal, maximal et moyenne des individus par rapport aux différentes occupations
data.groupby('occup').agg({'age':['min', 'max', 'mean']})
```

```python
# Genre qui passe plus de temps devant la télévision
data.groupby('sexe').agg({'heures.tv': 'mean'})
```

```python
plt.subplots(figsize = (14, 4), dpi = 150)
sns.boxplot(x = 'occup', y = 'poids', data = data)
plt.title("Distribution du poids selon l'occupation", pad = 16)
plt.xlabel('Occupation', fontsize = 12, labelpad = 16)
plt.show()
```

```python
plt.subplots(figsize = (14, 4), dpi = 150)
sns.boxplot(x = 'sport', y = 'age', data = data)
plt.title("Distribution de l'age selon la pratique du sport", pad = 16)
plt.xlabel('Sport', fontsize = 12, labelpad = 16)
plt.show()
```

```python
plt.subplots(figsize = (14, 4), dpi = 150)
sns.barplot(x = 'occup', y = 'poids', hue = 'sexe', errorbar=('ci', 0), data = data)
plt.title("Distribution du poids selon l'occupation et le sexe", pad = 16)
plt.xlabel('Occupation', fontsize = 12, labelpad = 16)
plt.show()
```

```python
# Utilisation de la fonction corr() de la librairie pandas
data[['age', 'poids']].corr()
```

```python
# Utilisation de la fonction corr() de la librairie pingouin
import pingouin as pg
pg.corr(data['age'], data['poids'])
```

```python
# Utilisation de la fonction pearsonr() de la librairie scipy
import scipy.stats as stat
stat.pearsonr(data['age'], data['poids'])
```
